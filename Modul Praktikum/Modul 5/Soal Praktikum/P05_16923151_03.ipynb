```python
import pandas as pd

data = pd.read_csv("P05-02.csv")
```

# 1. Histogram distribusi harga sewa per tahun dari rumah yang mengizinkan membawa hewan peliharaan, dengan bins 20

```python
req = (data.loc[(data["animal"] == "acept")])

req[["rent_amount"]].plot(kind="hist",bins=[0,1000,2500,5000,7500,10000,12500,15000,17500,20000],rwidth=0.8)
```

# 2. Pie chart perbandingan banyaknya masing-masing kota tempat rumah berada dari rumah dengan kamar mandi lebih dari 3.

```python
syarat1 = (data.loc[data["bathroom"] > 3])
syarat1["city"].value_counts().plot(kind = "pie")
```

# 3. Scatter plot dengan luas tanah sebagai x dan pajak bangunan sebagai y dari rumah yang tidak mengizinkan membawa hewan peliharaan

```python
syarat = (data.loc[(data["animal"] == "not acept")])
syarat.plot(kind="scatter", x="area", y="property_tax")
```

# 4. Horizontal bar chart menggambarkan jumlah rumah di setiap kota dengan jumlah kamar tidur lebih dari 3 dan tepat 2 kamar mandi.

```python
requirement = (data.loc[data["rooms"] >= 3])

requirement["city"].value_counts().plot(kind = "barh")


```
